```python
import os
import math

def convert_size(size_bytes):
   if size_bytes == 0:
       return "0B"
   size_name = ("B", "KB", "MB", "GB", "TB", "PB", "EB", "ZB", "YB")
   i = int(math.floor(math.log(size_bytes, 1024)))
   p = math.pow(1024, i)
   s = round(size_bytes / p, 2)
   return "%s %s" % (s, size_name[i])


png_img=Image.open('png_version.png')
h_im, l_im = png_img.size
im.save('./png_version.png', 'png') 
print("le format est:",png_img.format," ,les dimensions sont : (",h_im,",", l_im,") ,the type of image is:",png_img.mode)
size = os.path.getsize('png_version.png')
print("Taille de l'image :", convert_size(size))
```

```python
im_g = cv2.imread('exemple.jpg')
grayImage = cv2.cvtColor(im_g, cv2.COLOR_BGR2GRAY)
cv2.imwrite("Gray-image.jpg", grayImage)
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)
cv2.imwrite("noir-blanc-image.jpg", blackAndWhiteImage)
cv2.imshow('noir-blanc-image',blackAndWhiteImage)
cv2.imshow('Gray-image', grayImage)
cv2.imshow('original-image',im_g)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import random
def add_noise(img): 
# poivre et sel
    row , col ,d= img.shape 
    number_of_pixels = random.randint(300, 10000) 
    for i in range(number_of_pixels): 
        y_coord=random.randint(0, row - 1) 
        x_coord=random.randint(0, col - 1) 
        img[y_coord][x_coord] = 255
    number_of_pixels = random.randint(300 , 10000) 
    for i in range(number_of_pixels): 
        y_coord=random.randint(0, row - 1) 
        x_coord=random.randint(0, col - 1) 
        img[y_coord][x_coord] = 0
          
    return img 

im_g = cv2.imread('exemple.jpg')
poivre_sel_img = add_noise(im_g)
cv2.imwrite("poivre_sel_img.jpg", poivre_sel_img)

cv2.imshow('original-image',im_g)
cv2.imshow('poivre_sel_imgage',poivre_sel_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
